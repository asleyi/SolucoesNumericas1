$\bf 1)$ Usar o Python para plotar os campos de direção para a família de soluções de cada PVI da aula anterior. 

```python
# -----------------  Exercicio 1.5 ----------------
# ----------- EDO: y'=-[y(t)]^2+y(t), y(0)=1 ---------------------

import numpy as np
import matplotlib.pyplot as plt

# parametros
t0 , tb = 0, 3
y0 , yb = 1, 3
nt , ny = 10, 15

# dominio (t,y)
t = np. linspace (t0 ,tb ,nt)
y = np. linspace (y0 ,yb ,ny)
[T,Y] = np.meshgrid (t,y)

# EDO
dt = np.ones (T.shape)
dy = -(Y)**2 + Y

# campo
f = plt.figure (figsize = (6,5))
plt.quiver (T, Y, dt, dy, color ='b')
```

$\bf 2)$ Plotar as soluções particulares substituindo $c$.

```python
# ----------------- Exercicio 1.6 -----------------
# ----------- EDO: y'=-y+sin(t)+cos(t), y(0)=1 ---------------------

import numpy as np
import matplotlib.pyplot as plt

# parametros
t0 , tb = 0, 3
y0 , yb = -3, 3
nt , ny = 10, 15

# dominio (t,y)
t = np.linspace (t0 ,tb ,nt)
y = np.linspace (y0 ,yb ,ny)
[T,Y] = np.meshgrid (t,y)

# EDO
dt = np.ones (T.shape)
dy = -Y + np.sin(T) + np.cos(T)

# campo
f = plt.figure ( figsize =(6,5) )
plt.quiver (T, Y, dt, dy, color = 'k')

# solucoes particulares
y_part = lambda c: np.sin(t) + c*np.exp(-t)

for c in [-2., -1., 1., 2.]:
    aux = y_part(c)
    plt.plot (t, y_part (aux))
```

```python
# ----------------- Exercicio 1.6 -----------------
# ----------- EDO: y'=(y-y^2)/t, y(1)=2 ---------------------

import numpy as np
import matplotlib.pyplot as plt

# parametros
t0 , tb = 1, 3
y0 , yb = -2, 2
nt , ny = 10, 15

# dominio (t,y)
t = np.linspace (t0 ,tb ,nt)
y = np.linspace (y0 ,yb ,ny)
[T,Y] = np.meshgrid (t,y)

# EDO
dt = np.ones (T.shape)
dy = (Y-Y**2)/T

# campo
f = plt.figure ( figsize =(6,5) )
plt.quiver (T, Y, dt, dy, color = 'k')

# solucoes particulares
y_part = lambda c: t/(t+c)

for c in [-1, -0.5, 0.5, 1.]:
    aux = y_part(c)
    plt.plot (t, y_part (aux))
```
