$\bf Análise$ $\bf de$ $\bf estabilidade$

Repetir o experimento do exemplo para resolver o PVI 

$y'(t) = \lambda(y-sin(t)) + cos(t)$, $y(\dfrac {\pi}{4}) = \dfrac {1}{2}$ para $\lambda = 2$ e $\lambda = -2$. 

Comparar as soluções obtidas pelo método para $h = \dfrac {\pi}{10}$ e $h = \dfrac {\pi}{20}$ para cada valor de $\lambda$.

```python
# ------------- lambda = 2 -----------------
import numpy as np
import matplotlib.pyplot as plt
from numpy import *

def euler_expl (t0, tf, y0, h, fun):
    n = round ((tf-t0)/h + 1)
    t = linspace (t0, t0 +(n-1)*h,n)
    y = linspace (t0, t0 +(n-1)*h,n)
    y = zeros ((n ,))

    y[0] = y0
    

    for i in range (1,n):
        y[i] = y[i-1] + h*f(t[i-1], y[i-1])

    return (t,y)

f = lambda t,y: 2*(y-np.sin(t)) + np.cos(t)

# h = pi/10
plt.subplot (121)
for y0 in [0.5, 1, 2, 5]:
    (t, ynum ) = euler_expl (np.pi/4, 2, 1/2, np.pi/10, f)
    plt.plot (t, ynum )
    plt.title ('$h = pi/10 $')

# h = pi/20
plt.subplot (122)
for y0 in [0.5, 1, 2, 5]:
    (t, ynum ) = euler_expl (np.pi/4, 2, 1/2, np.pi/20, f)
    plt.plot (t, ynum )
    plt.title ('$h = pi/20 $')
```

```python
# ------------- lambda = -2 -----------------
import numpy as np
import matplotlib.pyplot as plt
from numpy import *

def euler_expl (t0, tf, y0, h, fun):
    n = round ((tf-t0)/h + 1)
    t = linspace (t0, t0 +(n-1)*h,n)
    y = linspace (t0, t0 +(n-1)*h,n)
    y = zeros ((n ,))

    y[0] = y0
    

    for i in range (1,n):
        y[i] = y[i-1] + h*f(t[i-1], y[i-1])

    return (t,y)

f = lambda t,y: -2*(y-np.sin(t)) + np.cos(t)

# h = pi/10
plt.subplot (121)
for y0 in [0.5, 1, 2, 5]:
    (t, ynum ) = euler_expl (np.pi/4, 2, 1/2, np.pi/10, f)
    plt.plot (t, ynum )
    plt.title ('$h = pi/10 $')

# h = pi/20
plt.subplot (122)
for y0 in [0.5, 1, 2, 5]:
    (t, ynum ) = euler_expl (np.pi/4, 2, 1/2, np.pi/20, f)
    plt.plot (t, ynum )
    plt.title ('$h = pi/20 $')
```
