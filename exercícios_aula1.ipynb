${\bf a)}$ Para cada PVI mostrar que a solução $y(t)$ dada satifaz a EDO e determinar o valor de $c$ para cada caso.

${\bf Caso 1:}$ $y' = -y + sin(t) + cos(t)$, $y(0) = 1$


Solução: $y(t) = sin(t) + ce^{-t}$ 

${\bf Resolvendo}$ ${\bf a}$ ${\bf EDO}$ $\bf (linear$ $\bf de$ $\bf primeira$ $\bf ordem): $ 

$dy = -y+sin(t)+cos(t)dt$


$dy+y=sin(t)+cos(t)dt$


Achando o fator integrante: $\mu(t)=e^{\int p(t)dt}= e^{\int 1dt} = e^{t}$


$\bf e^{t}\frac{dy}{dt} + e^{t}y$ = $e^{t}sin(t) + e^{t}cos(t)$


A parte em negrito é a derivada de $e^{t}y(t)$, ou seja, $\frac {d}{dt}(e^{t}y)$. 


Então temos: 


${\int \frac {d}{dt} (e^{t}y(t))}$ = ${\int e^{t}sin(t) + e^{t} cos(t)}$


$e^{t}y(t) = e^{t}sin(t) + c$


$y(t) = \dfrac {e^{^t}sin(t) + c}{e^{t}}$


$\bf y(t) = sin(t) + ce^{-t}$ como $\ Solução$ $Geral$



$\bf Aplicando$ $\bf o$ $\bf V.I.:$

$1 = sin(0) + ce^{0}$ => $c = 1$. Então temos $\bf y(t)$ = $\bf sin(t) + e^{-t}$ como $\ Solução$ $Particular$ para esse caso.

${\bf Mostrando}$ ${\bf que}$ ${\bf a}$ ${\bf Solução}$ $\bf {Geral}$ ${\bf satisfaz}$:

Como $y(t) = sin(t) + e^{-t}$ então $y'(t) = cos(t) - e^{-t}$. Substituindo na EDO teremos:

$=>$ $cos(t) - e^{-t} + sin(t) + e^{-t} = sin(t) + cos(t)$ 

$=>$ $cos(t) - cos(t) - e^{-t} + e^{-t} + sin(t) - sint (t) = 0$. Portanto, a solução satisfaz a EDO.

$\bf Caso2:$ $y' = \dfrac {(y-y^{2})}{t}$, $y(1) = 2$


Solução: $y(t) = \dfrac {t}{t+c}, t>0$

${\bf Resolvendo}$ ${\bf a}$ ${\bf EDO}$ $\bf (não-linear$ $\bf de$ $\bf primeira$ $\bf ordem): $ 


$\dfrac {dy}{(y-y^2)}$ = $\dfrac {dt}{t}$ $=>$ $(\dfrac {1}{y} - \dfrac {1}{y-1})dy = \dfrac {dt}{t}$. Aplicando a integral em ambos os lados e integrando obtemos:


$\int \dfrac {1}{y}dy - \int \dfrac {1}{y-1}dy = \int \dfrac {dt}{t}$ $=>$ $ln(y) - ln (y-1) = ln (t) + c$


Adicionando $ln (y+1)$ em ambos os lados obtemos:


$ln(y) - ln(y-1)$ + $\bf {ln(y+1)}$ = $ln(t)$ + $\bf {ln(y+1)}$ + $c$


$ln(y) = ln(t) + ln(y+1) + c$


$e^{ln(y)} = e^{ln(t) + ln(y+1) + c}$

$y = ct(y+1)$

$y = cty + ct$

$\bf y = \dfrac {t}{t+c}, t>0$ como $Solução$ $Geral$

<p> $\bf Aplicando$ $\bf o$ $\bf V.I.:$

$2 = \dfrac {1} {1+c}$ $=>$ $c = -\dfrac {1}{2}$. Então temos $y(t) = \dfrac {t}{t- \dfrac {1}{2}} = \dfrac {2t}{2t-1}$ como $Sol.$ $Particular$ para esse caso.

${\bf Mostrando}$ ${\bf que}$ ${\bf a}$ ${\bf Solução}$ ${\bf satisfaz}$:

Como $y(t) = \dfrac {t}{t-\dfrac {1}{2}}$, então $y'(t) = \dfrac {t+c - t(1+c)}{(t+c)^{2}} = 0$. Substituindo na EDO teremos:

$\dfrac {2}{(2t-1)^{2}} = \dfrac {(\dfrac {2t}{2t-1} - (\dfrac {2t}{2t-1})^{2})}{t}$

$=>$ $\dfrac {2}{(2t-1)^{2}} = -\dfrac {2}{(2t-1)^{2}}= 0$. Portanto, a solução satisfaz a EDO.


 $\bf Caso3:$ $y' = cos^{2}(y)$, $y(0)=\dfrac {\pi}{4}$
 
 $Solução: y(t) = arctan(t+c)$

${\bf Resolvendo}$ ${\bf a}$ ${\bf EDO:}$ 

$dy = cos^{2}(y)dt$. Aplicando a integral em ambos os lados e integrando, obtemos:

$\int \dfrac {dy}{cos^{2}(y)}$ = $\int dt$ 

$tan(y) = t+c$ $=>$ $\bf y = arctan (t+c)$ como $Solução$ $Geral$


<p> $\bf Aplicando$ $\bf o$ $\bf V.I.:$
    
$\dfrac {\pi}{4} = arctan(c)$. Se $arctan(x)=a$ $=>$ $x=tan(a)$, logo $c=tan (\dfrac{\pi}{4})$ $=>$ $c=1$. 

Temos então $\bf y(t) = arctan(t+1)$ como $Solução$ $Particular$  para esse caso.

${\bf Mostrando}$ ${\bf que}$ ${\bf a}$ ${\bf Solução}$ ${\bf satisfaz}$: 

${\bf Obs:}$ O cálculo da $tan$ foi feito em radianos.

Como $y(t) = arctan(t+1)$ então $y'(t) = \dfrac {1}{t^{2}+2t+2}$. Substituindo na EDO (na forma $y'-cos^{2}(y)=0$) obtemos:

$\dfrac {1}{t^{2}+2t+2} - cos^{2}(arctan(t+1))$

$=>$ $\dfrac {1}{t^{2}+2t+2}$ - $(\dfrac {\sqrt {1+(t+1)^{2}}}{1+(t+1)^{2}})^{2}$

$=>$ $\dfrac {1}{t^{2}+2t+2}$ - $\dfrac {1}{t^{2}+2t+2} = 0$. Portanto, a solução satisfaz a EDO.

$\bf c)$ Usar o critério da derivada para descutir o condicionamento do PVI

$y' = y - 5sin(t) -25cos^{2}(t)$, $y(0)=6$

$\bf Resposta:$ Um PVI geralmente é bem condicionado quando satisfaz o seguinte critério

$\dfrac {\partial f(t,y)}{\partial y}$ $\leq$ $1$, com $t_0$ $\leq$ $t$ $\leq$ $b$ 
  

Verificando esse critério no PVI dado:

$\dfrac {\partial (y-5sin(t)-25cos^{2}(t))}{\partial y} = 1$. Como $1$ $\geq$ $0$ então o PVI não é bem condicionado.
